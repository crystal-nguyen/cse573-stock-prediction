```python
import pandas as pd
import numpy as np

aapl_news = pd.read_csv('apple_news_stocks.csv')
amzn_news = pd.read_csv('amazon_news_stocks.csv')

aapl_tweets = pd.read_csv('aapl_tweets_stocks.csv')
amzn_tweets = pd.read_csv('amzn_tweets_stocks.csv')
```

```python
aapl_news.drop(columns=['Unnamed: 0','text'],inplace=True)
amzn_news.drop(columns=['Unnamed: 0','text'],inplace=True)
```

```python
aapl_news['Date'] = pd.to_datetime(aapl_news['publishedTime'].str.slice(0,10))
amzn_news['Date'] = pd.to_datetime(amzn_news['publishedTime'].str.slice(0,10))

aapl_news.drop(columns=['publishedTime'],inplace=True)
amzn_news.drop(columns=['publishedTime'],inplace=True)
```

```python
import yfinance as yf
from datetime import timedelta

aapl = yf.Ticker('AAPL')
amzn = yf.Ticker('AMZN')

aapl_min = min(aapl_news['Date'])
aapl_max = max(aapl_news['Date'])
aapl_max = aapl_max + timedelta(days=3)

amzn_min = min(amzn_news['Date'])
amzn_max = max(amzn_news['Date'])
amzn_max = amzn_max + timedelta(days=3)

aapl = aapl.history(start=aapl_min.date(), end=aapl_max.date())

amzn = amzn.history(start=amzn_min.date(), end=amzn_max.date())
```

```python
aapl.reset_index(inplace=True)
amzn.reset_index(inplace=True)

aapl = aapl[['Date','Close']]
amzn = amzn[['Date','Close']]

aapl['Next Close'] = aapl['Close'].shift(-1)
amzn['Next Close'] = amzn['Close'].shift(-1)

aapl.rename(columns={'Close':'AAPL Close','Next Close':'Next AAPL Close'},inplace=True)
amzn.rename(columns={'Close':'AMZN Close','Next Close':'Next AMZN Close'},inplace=True)
```

```python
aapl_news = aapl_news[['Date','title']].merge(aapl,on='Date')
aapl_news.rename(columns={'AAPL Close':'Close','Next AAPL Close':'Next Close'})

amzn_news = amzn_news[['Date','title']].merge(amzn,on='Date')
amzn_news.rename(columns={'AMZN Close':'Close','Next AMZN Close':'Next Close'})
```
